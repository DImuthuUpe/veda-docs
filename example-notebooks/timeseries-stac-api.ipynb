## Load and inspect one of the STAC items

We will use this list of STAC Items to generate the timeseries chart, but providing each STAC Item (which corresponds to a timestep) to the `/statistics` endpoint

```python
items = re.get(f"{STAC_API_URL}/collections/{collection}/items?limit=100").json()["features"]
items[0]
```

## Define a bounding box for which we will generate a timeseries

We will be using a bounding box over metropolitan france

```python
france_bounding_box = {
    "type": "Feature",
    "properties": {"ADMIN": "France", "ISO_A3": "FRA"},
    "geometry": {
        "type": "Polygon",
        "coordinates": [
            [
                [-5.183429, 42.332925],
                [8.233998, 42.332925],
                [8.233998, 51.066135],
                [-5.183429, 51.066135],
                [-5.183429, 42.332925],
            ]
        ],
    },
}



m = folium.Map(
    location=[40,0],
    zoom_start=3,
)

folium.GeoJson(france_bounding_box, name="France").add_to(m)
m

```

## Use `/cog/statistics` to get data for the desired area

First, we create a `generate_stats` function and then we call it with the bounding box defined for France.

```python
# the bounding box should be passed to the geojson param as a geojson Feature or FeatureCollection
def generate_stats(item, geojson):
    result = re.post(
        f"{RASTER_API_URL}/cog/statistics", 
        params={
            "url":item["assets"]["cog_default"]["href"]
        },
        json=geojson
    ).json()    
    return {
        **result["properties"], "start_datetime":item["properties"]["start_datetime"]
    }
   
```

## Generate and estimate time to generate statistics

This may take a couple minutes, due to the complexity of the shape we're requesting

```python
start = time.time()
stats = [generate_stats(item, france_bounding_box) for item in items]
end = time.time()
print(f"Elapsed time for bounding box (france): {round(end-start,2)} seconds. Items queried: {len(items)}")
```

## Inspect one result

```python
stats[1]
```

```python
dates = [dt.datetime.strptime(stat["start_datetime"], "%Y-%m-%dT%H:%M:%S") for stat in stats]

means = [stat["statistics"]["1"]["mean"] for stat in stats]

maxs = [stat["statistics"]["1"]["max"] for stat in stats]
mins = [stat["statistics"]["1"]["min"] for stat in stats]

std_devs = [stat["statistics"]["1"]["std"] for stat in stats]

upper_bounds = [m+s for (m,s) in zip(means, std_devs)]
lower_bounds = [m-s for (m,s) in zip(means, std_devs)]

fig = plt.figure(figsize=(20,10))

plt.plot(dates, means, 'black', label="Mean monthly NO2 values")

plt.fill_between(dates, upper_bounds, lower_bounds, facecolor="lightgray", interpolate=False, label="+/- one standard devation")

plt.plot(dates, mins, color='blue', linestyle="-", linewidth=0.5, label="Max monthly NO2 values")
plt.plot(dates, maxs, color='red', linestyle="-", linewidth=0.5, label="Min monhtly NO2 values")

plt.legend()
plt.title("NO2 Values in France (2016-2022)")
```

In this graph we can see the yearly cycles in NO2 values due to seasonal variations, as well as a slight downward slope in maximum NO2 values

### Comparison: multi vs single threaded approach

We can drastically reduce the time it takes to generate the timeseries, even with the complex AOI above, by taking advantage of the API's underlying infrastructure. The API is powered by [AWS Lambda](https://aws.amazon.com/lambda/), and executes each request in a separate instance, which makes it **highly** scalable. Since each statistics request is for a single timestamp, we can request statistics for multiple timesteps concurrently, and greatly reduce the amount of time needed. We will demonstrate this by using the `concurrent.futures` library: 

```python
from concurrent.futures import ThreadPoolExecutor 

# suspended function evaluation
def _generate_stats(item):
    return generate_stats(item, france_aoi)

start = time.time()
with ThreadPoolExecutor(max_workers=10) as executor: 
    aoi_stats = list(executor.map(_generate_stats, items))
end = time.time()
print(f"Elapsed time for a complex AOI (france): {round(end-start,2)} seconds. Items queried: {len(items)}")
```

By multi-threading the operation, we've reduced the runtime by a factor of almost 9 !
